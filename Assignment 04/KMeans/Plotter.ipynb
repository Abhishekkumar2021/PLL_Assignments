```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('data.csv')
true_labels = pd.read_csv('true_labels.csv')
generated_labels = pd.read_csv('generated_labels.csv')

plt.scatter(data['data_0'], data['data_1'], c=true_labels['true_labels'], cmap='viridis')
plt.show()

plt.scatter(data['data_0'], data['data_1'], c=generated_labels['generated_labels'], cmap='viridis')
plt.show()
```
